```python
import gseapy as gp
import numpy as np

from pathlib import Path
import pandas as pd

from matplotlib import rcParams
import matplotlib.pyplot as plt

rcParams["pdf.fonttype"] = 42

%config InlineBackend.figure_format = 'retina'
%matplotlib inline

DPI = 300 
FONTSIZE = 20  # 42
```

```python
gene_sign = "AP1_regulon_updated" 
receiver = "ipEMT"
```

```python
DIR2LOAD = Path("/data/BCI-CRC/nasrine/data/CRC/spatial/CRC_LM_VISIUM/CRC_LM_VISIUM_04_08_09_11/nichenet/concat_withWu2022/nichenet_microenvall_celltypes/intersect_cellphonedb")
```

```python
FIG2SAVE = DIR2LOAD.joinpath(f"figures/final/{gene_sign}_{receiver}/")
FIG2SAVE.mkdir(parents=True, exist_ok=True)
```

```python
msigdb_res = pd.read_excel(DIR2LOAD.joinpath(f"{gene_sign}{receiver}_common_ligands_gsea_filtered_significant.xls"),
              sheet_name="geneset_MSigDB_Hallmark_2020",
              index_col=0
             )

kegg_res = pd.read_excel(DIR2LOAD.joinpath(f"{gene_sign}{receiver}_common_ligands_gsea_filtered_significant.xls"),
              sheet_name="geneset_KEGG_2021_Human",
              index_col=0
             )
```

```python
msigdb_res
```

```python
kegg_res
```

```python
terms2highlight = ["Epithelial Mesenchymal Transition",
                   "Apoptosis",
                   "TNF-alpha Signaling via NF-kB",
                   "Inflammatory Response",
                   "Angiogenesis",
                   "Interferon Gamma Response",
                   "IL-6/JAK/STAT3 Signaling",
                   "Cytokine-cytokine receptor interaction"
                   "PI3K-Akt signaling pathway",
                   "MAPK signaling pathway",
                  ]
```

```python
res = pd.concat([msigdb_res, kegg_res])
```

```python
res = res[res["Term"].isin(terms2highlight)].copy()
```

```python
# compute log p-value for each term
res["-log10p"] = -np.log10(res["Adjusted P-value"])
res
```

```python
# sort by p-value
res.sort_values(by="-log10p", ascending=False, inplace=True)
res
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.barplot(data=res, x="-log10p", y="Term", palette="Blues_r")
ax.set_ylabel("")
ax.set_xlabel(u'-log\u2081\u2080(P)')
#ax.set_xlabel('$-log_{10}(P)$')
plt.savefig(FIG2SAVE.joinpath(f'{gene_sign}_{receiver}_gsea_barplot.pdf'), dpi=DPI,
            format="pdf", bbox_inches="tight",)
#ax.set_xticks([])
```
