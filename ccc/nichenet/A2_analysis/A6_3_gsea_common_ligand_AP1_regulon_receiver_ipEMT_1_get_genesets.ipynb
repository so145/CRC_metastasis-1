### run enrichment analysis

#"MSigDB_Hallmark_2020" # KEGG_2021_Human, GO_Biological_Process_2021, Reactome_2022

```python
geneset2use = ["MSigDB_Hallmark_2020", "KEGG_2021_Human"]
col = ["from", "to"] #"AP1_regulon_ligands"
```

```python
enr_dict = dict()

#gene2look = genes[col].values.tolist()
gene2look = list(set(genes[col].values.flatten()))

enr = gp.enrichr(gene_list=gene2look, 
                 gene_sets=geneset2use, 
                 organism="Human",
                 outdir=None,
                     #cutoff=0.05,
                )
enr_dict[gene_sign1] = enr.results
```

```python
enr_dict[gene_sign1]
```

### write results to file

```python
with pd.ExcelWriter(DIR2LOAD.joinpath(f'{gene_sign1}{receiver}_common_ligands_receptor_gsea.xls')) as writer:  
    for cluster in list(enr_dict.keys()):      
        enr_dict[cluster].to_excel(writer, sheet_name=f"region_{cluster}")
```

### Do a barplot of the relevant GO terms 
For each geneset:
* only select go terms that have adjusted pvalue < 0.01
* sort by adjusted pvalue < 0.01 

```python
final_res = {}
for s in geneset2use:
    df = enr_dict[f"{gene_sign1}"][enr_dict[f"{gene_sign1}"]["Gene_set"]==s].sort_values(by="Adjusted P-value")
    df = df[df["Adjusted P-value"]<0.01].sort_values(by="Adjusted P-value")
    final_res[s] = df
```

```python
with pd.ExcelWriter(DIR2LOAD.joinpath(f'{gene_sign1}{receiver}_common_ligands_receptor_gsea_filtered_significant.xls')) as writer:  
    for cluster in list(final_res.keys()):      
        final_res[cluster].to_excel(writer, sheet_name=f"geneset_{cluster}")
```
